## Task 1: Read Data From the 'woocommerce-product-export.csv'

```python
# import all required libraries
import pandas as pd                         # uses for load data, clean data, analysis data
from pathlib import Path                    # uses for handling file path
import matplotlib.pyplot as plt             # use to create visualization such as plot scatter chart, bar chart, hstogram chart, line chart
```

```python
# create file path
woocommerce_data = Path("woocommerce-product-export.csv")       # create a file path
data = pd.read_csv(woocommerce_data)                            # load the csv data
```

```python
# check if file exist
file_path = Path("woocommerce-product-export.csv") 
if file_path.is_file():
    print("File Exist - file is loading...")
    data = pd.read_csv(woocommerce_data)   
else: 
    print("No File Exists - Check file name or file path.")
```

```python
# load the file in to the DataFrame
df = pd.DataFrame(data)
```

## Print  the "total_profit" and “month_number” columns only

```python
#print only selected columns to inspect
print("Display Selected Columns:")
print(df[["total_profit", "month_number"]])
```

## Task 7: Read the total profit of all months and show it using the Bar plot.

```python
# We visualize the company profit per month
# generated using a bar chart.
#
# df.month_number
#     Provides the month number for the x-axis.
#
# df.total_profit
#     Provides the total profit values for the height of each bar.

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(df.month_number, df.total_profit)

plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

## Task 8: Read the total profit of all of the months

```python
# We visualize the company profit per month
# generated using a line chart.
#
# df.month_number
#     Provides the month number for the x-axis.
#
# df.total_profit
#     Provides the total profit values for the height of each bar.

import matplotlib.pyplot as plt
x = df.month_number
y = df.total_profit

plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Toatl Profit")

# Set plot colors and market
plt.plot(x, y, marker = 'o', markerfacecolor='black', linestyle='--', color="red", linewidth=3, label="Profit data of last year" )

# Enabling grid lines:
plt.legend(loc='lower right')
plt.show()
```

## Task 9: Print all of the product sales data

```python
# visualize all of the product sales data using a line chart.

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.plot(df.month_number, df.facecream, '-', color='blue', marker="o", label="Face Cream Sales Data")
plt.plot(df.month_number, df.facewash, '-.', color='orange', marker="o", label="Face Wash Sales Data")
plt.plot(df.month_number, df.toothpaste, '-', color='green', marker="o", label="ToothPaste Sales Data")  # Plot job rate data
plt.plot(df.month_number, df.bathingsoap, '-', color='red', marker="o", label="Bathingsoap Sales Data")
plt.plot(df.month_number, df.shampoo, '-', color='purple', marker="o", label="Sahmpoo Sales Data")
plt.plot(df.month_number, df.moisturizer, '-', color='brown', marker="o", label="Moisturizer Sales Data")

plt.ylabel("Sales units in number")
plt.xlabel("Month Number")
plt.title("Sales Data")
plt.legend()

plt.show()
```

## Task 10: Read “bathingsoap” sales data for each month

```python
# visualize bathingsoap sales data for each month using a schatter chart.
#
# df.month_number
#     Provides the month number for the x-axis.
#
# df.bathingsoap
#     number of unit sold for y axis.
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(df.month_number, df.bathingsoap, label="Bathingsoap Sales Data")

plt.ylabel("Number of Unit Sold")
plt.xlabel("Month Number")
plt.title("bathingsoap Sales data")

plt.grid(linestyle='--')
plt.legend()
plt.show()
```

## Task 2: Average weight against the average height 

```python
# visualize the average weight againist the average height using line chart.
#
# height
#     Provides the height number for the x-axis.
#
# weight
#     Provides the weight values for the y-axis.

height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]

# Set plot colors and market
plt.plot(weight, height, marker = 'o', ms= 10, lw=3, linestyle='dashdot', color="green")
plt.title("Average weight with respect to average height")
plt.xlabel("Weight in kg")
plt.ylabel("Height in cm")
plt.show()
```
